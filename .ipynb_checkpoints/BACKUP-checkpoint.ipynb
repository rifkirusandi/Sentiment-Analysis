```python
import requests
from bs4 import BeautifulSoup

URL = 'https://www.tripadvisor.com/Attraction_Review-g469404-d3780963-Reviews-Double_Six_Beach-Seminyak_Kuta_District_Bali.html#REVIEWS'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
review = soup.find('div', class_='_2rspOqPP')
```

```python
for review in soup.find('div', class_='_2rspOqPP'):
    review_user = review.find('div', class_='_1svEu5jc _14_buatE').span.a.text
    review_country = review.find('div', class_='DrjyGw-P _26S7gyB4 NGv7A1lw _2yS548m8 _2cnjB3re _1TAWSgm1 _1Z1zA2gh _2-K8UW3T _1dimhEoy').span.text
    review_content = review.find('div', class_='DrjyGw-P _26S7gyB4 _2nPM5Opx').span.text
    review_date = review.find('div', class_='_3JxPDYSx').text

    print(review_user)
    print(review_country)
    print(review_date)
    print(review_content)
```
